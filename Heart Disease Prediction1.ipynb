```python
#Python have pandas library to create a dummy variable ie; pd.get_dummies()


dataset=pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
dataset
```

# Feature Scalling

Scale the values in the dataset to be a value in between 0 and 1

```python
from sklearn.preprocessing import StandardScaler
StandardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = StandardScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

```python
# Split the dataset into features data and target data

y = dataset['target']
X = dataset.drop(['target'], axis=1)
```

```python
#Split the dataset again, into 80% training dataset and 20% twsting dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=1)
```

# Implementing K-Nearest Neighbors(KNN) Algorithm

```python

from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

knn_scores = []
for k in range(1,30):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,y,cv=10)
    knn_classifier.fit(X_train, y_train)
    prediction = knn_classifier.predict(X_test)
    knn_scores.append(score.mean())
    
```

```python

plt.plot([k for k in range(1,30)], knn_scores, color = 'Green')
for i in range(1,30):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
    plt.xlabel('Number of Neighbors (k)')            
    plt.ylabel('Scores')
    
    plt.title('K Neighbors Classifier scores for different k values')
    plt.rcParams['figure.figsize']=(50,30)
```

# Testing the model accuracy  in a training dataset

```python
model=knn_classifier
model.score(X_train, y_train)
print("Accuracy on the training dataset: ", model.score(X_train, y_train))
```

# Testing the model accuracy  in a test dataset

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
print(metrics.classification_report(y_test, model.predict(X_test)))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score = accuracy_score(y_test, model.predict(X_test))
print("Model Accuracy on test dataset: ", accuracy_score)
```
